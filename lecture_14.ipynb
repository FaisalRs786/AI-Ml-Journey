# working with seaborn plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Titanic.csv')
df.head(10)
```

**THERE ARE DIFFFERENT COLOR PALETTES IN SEASBORN**

```python
# categorical color palette
sns.palplot(sns.color_palette('bright'))
plt.show()

current_color1 = sns.color_palette()
sns.palplot(current_color1)
plt.show()

sns.palplot(sns.color_palette('pastel'))
plt.show()
```

```python
# contenious color
sns.palplot(sns.color_palette('Blues',n_colors=8))
plt.show()
```

```python
# diverging color palette
sns.palplot(sns.color_palette('BrBG',20))
plt.show()

# for color blinders
sns.palplot(sns.color_palette('colorblind'))
```

# Distribution plot displot()

```python
# histplot()

sns.histplot(df['Age'], kde = False)
plt.show()

# kdeplot
sns.histplot(df['Age'], kde = True)
plt.show()

# distplot()
sns.displot(df['Age'], kde = True)
plt.show()
```

# RELATIONAL PLOT

```python
# relplot()
sns.relplot(x = 'Survived' , y= 'Age' ,data = df, col = 'Sex', hue = 'Pclass', style = 'Pclass')
plt.show()

# scatterplot()
sns.scatterplot(x = 'Sex', y = 'Age', data = df, hue = 'Pclass')
plt.show()

# lineplot()
sns.lineplot(x = 'Sex', y = 'Survived', data = df)
plt.show()
```

# Categorical Plot

```python
# barplot()
sns.barplot(x = 'Survived', y = 'Pclass', data = df, hue = 'Sex')
plt.yticks([0,1,2])
plt.show()

# striplot()\
sns.stripplot(x = 'Survived', y = 'Age', data = df, hue = 'Sex')
plt.show()

# swarmplot()
sns.swarmplot(x = 'Survived', y = 'Age', data = df, hue = 'Sex')
plt.show()

# boxplot()
sns.boxplot(x = 'Survived', y = 'Age', data = df, hue = 'Sex')
plt.show()

# violinplot()
sns.violinplot(x = 'Survived', y = 'Age', data = df, hue = 'Sex')
plt.show()

# countplot()
sns.countplot(x = 'Survived', data = df, hue = 'Sex')
plt.show()

# pointplot()
sns.pointplot(x = 'Survived', y = 'Age', data = df, hue = 'Sex')
plt.show()
```

# REGRESSION PLOT

```python
# lmplot()
sns.lmplot(x = 'Survived', y = 'Age', data = df, hue = 'Sex')
plt.show()

# regplot()
sns.regplot(x = 'Survived', y = 'Age', data = df)
plt.show()

# residplot()
sns.residplot(x = 'Survived', y = 'Age', data = df)
plt.show()
```

# MATRIX PLOT

```python
# heatmap
sns.heatmap(df.select_dtypes(include = ['number']).corr(), annot = True, cmap = 'YlGnBu')
plt.show()

# clustermap()
sns.clustermap(df.select_dtypes(include = ['number']).corr(), annot = True)
plt.show()
```

# MULTIPLE PLOT

```python
# Fcetgrids()

q = sns.FacetGrid(df , col = 'Survived', row= 'Pclass',hue = 'Embarked')

q.map(plt.hist, 'Sex')
plt.legend()
plt.show()
```
