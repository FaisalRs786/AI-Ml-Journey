FIRST WORK ON DATA SET TITANIC  

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
```

```python
df = pd.read_csv('/content/Titanic.csv')
df.head()
```

HERE FIRST I DROP THE TICKET COLUMN AND NAME COLUMN AND PASSENGERID COLUMN AS :

```python
df.drop(['Ticket','Name','PassengerId'],axis=1,inplace=True)
df
```

```python
df.info()
```

now we observe that our two columns are not in mumerical that are in object/string data columns are [ Sex and Embarker ]
> so we first convert these two columns dtype is categorical as:

```python
obj_var = ['Sex','Embarked']
for var in obj_var:
  df[var] = df[var].astype('category')
df.info()
```

now we can see our columns dtype is changed in to category

> now we convert these category column data into numerical or int using get_dummies function as:

```python
sex_var = pd.get_dummies(df['Sex'] , drop_first = True , dtype = int)
embarked_var = pd.get_dummies(df['Embarked'], drop_first = True , dtype = int)
```

now our sex and embarkeed dummy column are ready

```python
print(sex_var)
print(embarked_var)
```

now we put these tho columns in data set and remove previous Sex and Embarked columns from data set as:

```python
df = pd.concat([df, sex_var,embarked_var],axis=1)
df.drop(['Sex','Embarked'],axis=1,inplace=True)
df.head()
```

```python
df.head()
```

**now out data is ready to give to model to train**

# split data

```python
x = df.drop('Survived',axis=1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=44)
```

# DECISIONTREECLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

> model evaluation

```python
dtr_pred = dtc.predict(x_test)

# Accuracy
from sklearn.metrics import accuracy_score
dtr_accuracy = round(accuracy_score(y_test,dtr_pred)*100,2)
print("decision tree accuracy ",dtr_accuracy,"%")
```

# RANDOMFORESTCLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
```

> model evaluation

```python
rfc_pred = rfc.predict(x_test)

# Accuracy
from sklearn.metrics import accuracy_score
rfc_accuracy = round(accuracy_score(y_test,rfc_pred)*100,2)
print("random forest accuracy ",rfc_accuracy,"%")
```

#GRADIENTBOOSTINGCLASSIFIER

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
gbc.fit(x_train,y_train)
```

> model evaluation

```python
gbc_pred = gbc.predict(x_test)

# Accuracy
from sklearn.metrics import accuracy_score
gbc_accuracy = round(accuracy_score(y_test,gbc_pred)*100,2)
print("gradient boosting accuracy",gbc_accuracy,"%")
```
